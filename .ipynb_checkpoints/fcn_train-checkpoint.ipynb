```python
num_channels = 3

# USER DEPENDENT PATHS. Change these for your purposes.
weights_path = '../models/vgg16_tf.h5'          # not used
raw_data_dir = 'data/raw'                    # directory containing raw images
label_data_dir = 'data/label'                # directory containing labeled images
train_list = 'data/train.txt'                # list of subset of raw images to use for training
val_list = 'data/val.txt'                    # validation set. (not currently used)
train_raw_npy = 'data/train_raw.npy'         # path to save the numpy array of training images
train_label_npy = 'data/train_label.npy'     # path to save the numpy array of labelled images with labels 0:bg, 1:mito, 2:er, 3:nm, 4:pm
checkpoint_file = 'mynet.hdf5'                  # intermediate weights files

img_width = 256                                 # size of training images
img_height = 256                                # size of training images
input_size = (img_width, img_height)       
zoom = 1 # downsample factor
num_classes = 5                                 # number of labels

batch_size = 8   # make this as big a memory allows


subtract_mean = 112
nb_epoch = 25
```

```python
def argmax(matrix):
    return K.argmax(matrix,axis=1)

def get_class_weight(weights=[]):
    # percentages 4,1,1,3,91
    m = len(weights)
    w_array = 1e6*np.ones((m,m))
    # true, predicted
    #w_array[2,:] = 1e1
    #w_array[:,4] = 1e2  # if not bg but predicted bg
    w_array[0,:] *= 1.0/4
    w_array[3,:] *= 1.0/3
    w_array[4,:] *= 1.0/91  # if bg, but predicted else
    #for i in range(4):
    	#w_array[i,i] = 1e1
    return w_array

# from keras issue 2115
def weighted_crossentropy(y_true, y_pred, class_weight):
    print (y_pred,y_true)
    nb_cl = len(class_weight)

    final_mask = K.zeros_like(y_pred[:,:,0])  # shape 65000
    y_pred_max = K.max(y_pred, axis=2)
    y_pred_max = K.expand_dims(y_pred_max, 2)
    y_pred_max_mat = K.equal(y_pred, y_pred_max)  #one hot of pred. shape 5,256,256
    for c_p, c_t in product(range(nb_cl), range(nb_cl)):
        # class-weight[i,j] * y_pred_max_mat[:,j] * y_true[:,i]
        final_mask += (K.cast(class_weight[c_t, c_p],K.floatx()) * K.cast(y_pred_max_mat[:,:, c_p] ,K.floatx())* K.cast(y_true[:,:, c_t],K.floatx()))
   
    #y_pred=K.permute_dimensions(y_pred, (0,2,3,1))     
    #y_true=K.permute_dimensions(y_true, (0,2,3,1))     
    crossentropy = K.categorical_crossentropy(y_pred, y_true) 
    return crossentropy * final_mask


# create data given directory containing all images and list
def create_data(data_path, list_file, npy_file, zoom=1,is_label=False):
    with open(list_file,'r') as f:
        names = f.read().split('\n')[:-1]   # exclude last empty line
    num_images = len(names) 

    if is_label:
        datatype = np.uint8
        imgs = np.ndarray((num_images, img_height* img_width,num_classes), dtype=datatype)
    else:
        datatype = np.float32
        imgs = np.ndarray((num_images, img_height, img_width, num_channels), dtype=datatype)

#     if num_channels == 1:
#         imread_mode = cv2.IMREAD_GRAYSCALE
#     elif num_channels == 3:
#         imread_mode = cv2.IMREAD_COLOR

    print('-'*30)
    print('Creating training images...')
    print('-'*30)


    i = 0
    for name in names:
        if is_label:
            img = Image.open(os.path.join(data_path, name))
            img = 4-img   # I flipped the labels so that bugs are more noticeable
            #img = img.resize(img_height*img_width) #cv2.resize(img,(img_height,img_width),interpolation=cv2.INTER_NEAREST)
            one_hot = np.zeros((num_classes,img_height,img_width))
            x = np.arange(img_width)
            y = np.arange(img_height)
            x,y = np.meshgrid(x,y)
            one_hot[img,y,x] = 1
            img = one_hot.reshape(num_classes,img_height*img_width)
            img = img.transpose(1,0)
        else:
            img = Image.open(os.path.join(data_path, name)) #cv2.imread(os.path.join(data_path, name),imread_mode)
        if num_channels == 1:
            img = np.expand_dims(img,2) 

        print(img)
        plt.plot(img)
        img = np.array([img])
        imgs[i] = img

        if i % 100 == 0:
            print('Done: {0}/{1} images'.format(i, num_images))
        i += 1

    if not is_label and K.image_dim_ordering() == 'th':
        imgs = imgs.transpose(0,3,1,2)
    print('Loading done.')
    np.save(npy_file, imgs)
    print('Saving to {0} files done.'.format(npy_file))


def load_data(raw, label):
    imgs_raw = np.load(raw)
    imgs_label = np.load(label)
    return imgs_raw, imgs_label

def load_net(weights_path,num_classes=num_classes,input_size=input_size):
    return resnet(None,num_classes=num_classes,input_size=input_size)

def identity_block(input_tensor, kernel_size, filters, stage, block,atrous_rate=(1,1)):
    '''The identity_block is the block that has no conv layer at shortcut
    # Arguments
        input_tensor: input tensor
        kernel_size: defualt 3, the kernel size of middle conv layer at main path
        filters: list of integers, the nb_filters of 3 conv layer at main path
        stage: integer, current stage label, used for generating layer names
        block: 'a','b'..., current block label, used for generating layer names
    '''
    nb_filter1, nb_filter2, nb_filter3 = filters
    if K.image_dim_ordering() == 'tf':
        bn_axis = 3
    else:
        bn_axis = 1
    conv_name_base = 'res' + str(stage) + block + '_branch'
    bn_name_base = 'bn' + str(stage) + block + '_branch'

    x = Convolution2D(nb_filter1, 1, 1, name=conv_name_base + '2a')(input_tensor)
    x = BatchNormalization(axis=bn_axis, name=bn_name_base + '2a')(x)
    x = Activation('relu')(x)

    x = AtrousConvolution2D(nb_filter2, kernel_size, kernel_size, atrous_rate=atrous_rate,
                      border_mode='same', name=conv_name_base + '2b')(x)
    x = BatchNormalization(axis=bn_axis, name=bn_name_base + '2b')(x)
    x = Activation('relu')(x)

    x = Convolution2D(nb_filter3, 1, 1, name=conv_name_base + '2c')(x)
    x = BatchNormalization(axis=bn_axis, name=bn_name_base + '2c')(x)

    x = merge([x, input_tensor], mode='sum')
    x = Activation('relu')(x)
    return x

def conv_block(input_tensor, kernel_size, filters, stage, block, strides=(1, 1),atrous_rate=(1,1)):
    '''conv_block is the block that has a conv layer at shortcut
    # Arguments
        input_tensor: input tensor
        kernel_size: defualt 3, the kernel size of middle conv layer at main path
        filters: list of integers, the nb_filters of 3 conv layer at main path
        stage: integer, current stage label, used for generating layer names
        block: 'a','b'..., current block label, used for generating layer names
    Note that from stage 3, the first conv layer at main path is with subsample=(2,2)
    And the shortcut should have subsample=(2,2) as well
    '''
    nb_filter1, nb_filter2, nb_filter3 = filters
    if K.image_dim_ordering() == 'tf':
        bn_axis = 3
    else:
        bn_axis = 1
    conv_name_base = 'res' + str(stage) + block + '_branch'
    bn_name_base = 'bn' + str(stage) + block + '_branch'

    x = Convolution2D(nb_filter1, 1, 1, subsample=strides,
                      name=conv_name_base + '2a')(input_tensor)
    x = BatchNormalization(axis=bn_axis, name=bn_name_base + '2a')(x)
    x = Activation('relu')(x)

    x = AtrousConvolution2D(nb_filter2, kernel_size, kernel_size, border_mode='same',
                      atrous_rate=atrous_rate, name=conv_name_base + '2b')(x)
    x = BatchNormalization(axis=bn_axis, name=bn_name_base + '2b')(x)
    x = Activation('relu')(x)

    x = Convolution2D(nb_filter3, 1, 1, name=conv_name_base + '2c')(x)
    x = BatchNormalization(axis=bn_axis, name=bn_name_base + '2c')(x)

    shortcut = Convolution2D(nb_filter3, 1, 1, subsample=strides,
                             name=conv_name_base + '1')(input_tensor)
    shortcut = BatchNormalization(axis=bn_axis, name=bn_name_base + '1')(shortcut)

    x = merge([x, shortcut], mode='sum')
    x = Activation('relu')(x)
    return x


def resnet(weights_path,num_classes=num_classes,input_size=input_size):
    
    # for predictions with variable size    

    #using tf dim_ordering
    if K.image_dim_ordering() == 'tf':
        input_shape = (input_size[0],input_size[1],num_channels)
        bn_axis = 3
    else:
        input_shape = (num_channels,input_size[0],input_size[1])
        bn_axis = 1
    
    img_input = Input(shape=input_shape)
    num_outputs = [32,32,64]
    x = ZeroPadding2D((3, 3))(img_input)
    x = Convolution2D(num_outputs[0], 7, 7, subsample=(1, 1), name='conv1')(x)  # keras 1 API
    #x = Conv2D(num_outputs[0], (7, 7), strides=(1, 1), name='conv1')(x)  # keras 2 API syntax
    x = BatchNormalization(axis=bn_axis, name='bn_conv1')(x)
    x = Activation('relu')(x)
    #x = MaxPooling2D((3, 3), strides=(2, 2))(x)
    
    atrous_rate = (2,2)
    x = conv_block(x, 3, num_outputs, stage=2, block='a'    , atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=2, block='b', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=2, block='c', atrous_rate=atrous_rate)

    x = conv_block(x, 3, num_outputs, stage=3, block='a'    , atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=3, block='b', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=3, block='c', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=3, block='d', atrous_rate=atrous_rate)

    x = conv_block(x, 3, num_outputs, stage=4, block='a'    , atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=4, block='b', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=4, block='c', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=4, block='d', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=4, block='e', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=4, block='f', atrous_rate=atrous_rate)

    x = conv_block(x, 3, num_outputs, stage=5, block='a'    , atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=5, block='b', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=5, block='c', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=5, block='d', atrous_rate=atrous_rate)

    x = conv_block(x, 3, num_outputs, stage=6, block='a'    , atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=6, block='b', atrous_rate=atrous_rate)
    x = identity_block(x, 3, num_outputs, stage=6, block='c', atrous_rate=atrous_rate)

    x = conv_block(x, 3, [32,32,num_classes], stage=7, block='a', atrous_rate=atrous_rate)
    #x = AveragePooling2D((7, 7), name='avg_pool')(x)
   
    #x = Convolution2D(num_classes, 1, 1, activation='relu',name='fc')(x)
    x = Reshape((input_size[0]*input_size[1],num_classes))(x)
    x = Activation('softmax')(x)
    
    model = Model(img_input, x)

    if weights_path:
        model.load_weights(weights_path)

    model.summary()

    return model


def train(finetune=False,lr=1e-3):
    print('-'*30)
    print('Loading and preprocessing train data...')
    print('-'*30)
   
    # Create training data and save as numpy array. Only need to do this once per training dataset
    create_data(raw_data_dir,train_list,train_raw_npy,zoom=zoom)
    create_data(label_data_dir,train_list,train_label_npy,zoom=zoom,is_label=True)


    # Load training data from numpy array.
    train_raw, train_label = load_data(train_raw_npy,train_label_npy)
    train_raw = train_raw.astype('float32') - subtract_mean
    train_label = train_label.astype('float32')

    print('-'*30)
    print('Creating and compiling model...')
    print('-'*30)

    # load the model architecture
    model = load_net(weights_path,num_classes=num_classes,input_size=input_size)

    #if finetune:
   	    #print("Finetuning using", checkpoint_file)
        #model.load_weights(checkpoint_file)
    model_checkpoint = ModelCheckpoint(checkpoint_file, monitor='loss' )

    print('-'*30)
    print('Fitting model...')
    print('-'*30)

    # this randomly flips / rotates the images for invariance. Not sure if rotation works.
    datagen = ImageDataGenerator(
            rotation_range=90,
            vertical_flip=True,
            horizontal_flip=True,
            )

    # class weighting. Train with class balancing. Then finetune with all weights set to 1.
    class_weight = get_class_weight(np.ones(num_classes))

    # define custom loss
    ncce = partial(weighted_crossentropy, class_weight=class_weight)
    ncce.__name__ ='weighted_crossentropy'

   
    # Compile the model. The parameters for the optimizer can be changed... but these are standard. 
    model.compile(loss=ncce,
                  optimizer=optimizers.Adam(lr=lr, beta_1=0.9, beta_2=0.999, epsilon=1e-08, decay=0.0),
                  metrics=['accuracy'] )

    
    # Train
    for i in range(2):
        K.set_value(model.optimizer.lr, lr)
        model.fit_generator(datagen.flow(train_raw, train_label, batch_size=batch_size, shuffle=True),
                samples_per_epoch=len(train_raw), nb_epoch=nb_epoch, verbose=1, callbacks=[model_checkpoint,TensorBoard()])
        lr = lr *.1
```

```python
train()
```
